#**Estudo de Caso**

Temos uma lista de notas da turma de Python 101 compostas dos resultados das avaliações N1, N2, Sub formatada em um arquivo csv.

Nossa missão é calcular a Média Final e exibir o resultado da turma por aluno e depois a Méda Global da Turma.

---
**Segue forma de cálculo da métrica:**

**MF** = (((N1 x 0.4) + (N2 x 0.6)) /(N1+N2)).

**Caso não tenha atingido a Média com a N1 e N2:**

**MF com Sub** = (((N1 x 0.4) + (N2 x 0.6)) + Sub /(N1+N2+Sub))

--- 

A Sub só é aplicada quando o aluno não atinge a média com a N1 e N2.

**O Aluno é aprovado em caso de Média Final Maior ou Igual a 7. Menor que 7 o aluno é encaminhado para DP.**




**Layout do Arquivo**

matricula,nome,n1,n2,sub

**Dicionário de Dados**

Dicionário de Dados é um artefato de arquitetura de dados que documenta a estrutura de uma entrada de dados, criando um "contrato" entre utilizadores da estrutura.


|Campo |Tipo |Posição
|:-----: |:----: |:-------:
|matricula|Inteiro|1
|nome|Alfanumérico|2
|n1|Real|3
|n2|Real|4
|sub|Real|5

```python
import pandas as pd 
import io 
  
df = pd.read_csv(io.BytesIO(uploaded['notas.csv'])) 
print(df)
```

```python
total_alunos = 0
nota_total = 0
for aluno in df:

  #Alocando as informações em variaveis. 
  codigo_aluno = df.loc[total_alunos,"codigo_aluno"]
  nome_aluno = df.loc[total_alunos,"nome_aluno"]
  n1 = df.loc[total_alunos,"n1"]
  n2 = df.loc[total_alunos,"n2"]
  sub = df.loc[total_alunos,"sub"]

  #Calculando as médias por aluno
  media_ponderada_nota = (((n1*0.4)+(n2*0.6))/(n1+n2))*10

  #Vendo se foi aprovado ou não
  if media_ponderada_nota <= 7:
    media_com_sub = ((n1 * 0.4) + (n2 * 0.6) + sub) /(n1+n2+sub)*10
    if media_com_sub >= 7:
      print(f"Aprovado com Sub, \nAluno: {nome_aluno} \nNota Final:  {media_com_sub:.2f}")
    else: 
      print(f"Reprovado, \nAluno: {nome_aluno} \nNota: {media_com_sub:.2f}\n")



  #print(codigo_aluno)
  #print(nome_aluno)
  #print(n1)
  #rint(n2)
  #print(sub)
  #print(media_ponderada_nota)
  total_alunos += 1

#Média total da turma
soma_n1 = df["n1"].sum()
soma_n2 = df["n2"].sum()
media_da_turma = ((soma_n1 * 0.4) + (soma_n2 * 0.6))/(soma_n1 + soma_n2)*10
print(f"\nMédia da turma: {media_da_turma:.2f}")
print(f"\nTotal de alunos: {total_alunos}")

#MF = (((N1 x 0.4) + (N2 x 0.6)) /(N1+N2)).
```

```python
### Dica Augusto Firmo - Slack da Tera ### 

#Calcular a média ponderada diretamento do dataframe
media_ponderada = (df['n1']*0.4 + df['n2']*0.6)/(df['n1']+df['n2'])*10

#Podemos adicionar a média ponderada ao dataframe
df['media_ponderada'] = media_ponderada

#Como todos possuem notas de sub, podemos calcular sua média com sub para todos os alunos e realizar o tratamento na impressão
media_sub = ((df['n1']*0.4) + (df['n2']*0.6) + df['sub'])/(df['n1']+df['n2']+df['sub'])*10
df['media_sub'] = media_sub

print(df)
print("\n")

for i in range(len(df)):
    if df['media_ponderada'][i] < 7:
        if df['media_sub'][i] >= 7:
            print(f'Aluno: {df["nome_aluno"][i]}\nSituação: Aprovado com sub\nNota final: {df["media_sub"][i]:.2f}')
        else:
            print(f'Aluno: {df["nome_aluno"][i]}\nSituação: Reprovado com sub\nNota final: {df["media_sub"][i]:.2f}')
    else:
        print(f'Aluno: {df["nome_aluno"][i]}\nSituação: Aprovado sem sub\nNota final: {df["media_ponderada"][i]:.2f}')
    
    print('\n')

print("Numero de alunos:")
print(df["nome_aluno"].count())


media_turma = ((df["n1"].sum() * 0.4) + (df["n2"].sum() * 0.6))/(df["n1"].sum() + df["n2"].sum())*10
print("Média da turma:")
print(f'{media_turma:.2f}')
```
