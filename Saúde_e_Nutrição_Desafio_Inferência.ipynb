## Idade e Sintomas Depressivos

## Os 5 grupos de depressão diferem quanto a idade? 

Teste ANOVA (F): Compara as médias de mais de dois grupos


*   H0: Não há diferença entre as médias dos grupos quando a idade
*   H1: Há pelo menos uma diferença entre as médias dos grupos quanto a idade



```python
from scipy.stats import f_oneway

df_aux = df[["PHQ_GRP", "RIDAGEYR"]].dropna()

stat, p = f_oneway(df_aux[(df_aux.PHQ_GRP == 0)]["RIDAGEYR"],
                   df_aux[(df_aux.PHQ_GRP == 1)]["RIDAGEYR"],
                   df_aux[(df_aux.PHQ_GRP == 2)]["RIDAGEYR"],
                   df_aux[(df_aux.PHQ_GRP == 3)]["RIDAGEYR"],
                   df_aux[(df_aux.PHQ_GRP == 4)]["RIDAGEYR"],
                   )

print('stat=%.3f, p=%.3g' % (stat,p))
```

Conclusão: Os três grupos não diferem quanto a idade

## Os Três Grupos de depressão diferem quanto ao tempo dedicado a exercício Físico, por semana? 

Teste ANOVA (F): Compara as médias de mais de dois grupos


*   H0: Não há diferença entre as médias dos grupos quando a média de tempo gasto com exercicio fisico
*   H1: Há pelo menos uma diferença entre as médias dos grupos quando a média de tempo gasto com exercicio fisico

```python
df_aux = df[["PHQ_GRP", "PAG_MINW"]].dropna()

stat, p = f_oneway(df_aux[(df_aux.PHQ_GRP == 0)]["PAG_MINW"],
                   df_aux[(df_aux.PHQ_GRP == 1)]["PAG_MINW"],
                   df_aux[(df_aux.PHQ_GRP == 2)]["PAG_MINW"],
                   df_aux[(df_aux.PHQ_GRP == 3)]["PAG_MINW"],
                   df_aux[(df_aux.PHQ_GRP == 4)]["PAG_MINW"],
                   )

print('stat=%.3f, p=%.3g' % (stat,p))
```

Conclusão: Os três grupos diferem quanto o tempo praticado de exercicio fisico

Quais grupos diferem entre si? 

Para responder a essa pergunta, faremos o teste post-hoc de Tukey

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(df_aux["PAG_MINW"],
                          df_aux["PHQ_GRP"],
                          alpha=0.05)

print(tukey)
```

## Genero e sintomas depressivos

A proporção de pessoas em cada grupo de depressão varias de acordo com Genero? 

Teste qui-quadrado: avalia a associação entre duas variaveis categorias. 

Hipoteses do teste qui-quadrado



*   H0: Não há associação entre gênero e a presença de sintomas depressivos
*   H1: Há associação entre gênero e a presença de sintomas depressivos



```python
from scipy.stats import chi2_contingency

crosstab = pd.crosstab(df.replace(label_quali)["RIAGENDR"],
                       df.replace(label_quali)["PHQ_GRP"]
                       )

stat, p, df, expected = chi2_contingency(crosstab)
print('stat = %.3f, p = %.3f' % (stat,p))
```

H1: Há associação entre gênero e a presença de sintomas depressivos
